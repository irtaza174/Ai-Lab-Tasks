```python
def bfs_without_queue(graph, start):
    visited = set()
    list_queue = [start]     
    print("BFS Traversal:", end=" ")
    while list_queue:
        node = list_queue.pop(0)  

        if node not in visited:
            print(node, end=" ")
            visited.add(node)

            for neighbor in graph[node]:
                if neighbor not in visited:
                    list_queue.append(neighbor)


graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}
bfs_without_queue(graph, 'A')
```

```python
from collections import deque

def bfs_with_queue(graph, start):
    visited = set()                 
    queue = deque([start])          
    print("BFS Traversal:", end=" ")
    while queue:
        node = queue.popleft()     

        if node not in visited:
            print(node, end=" ")
            visited.add(node)
            for neighbor in graph[node]:
                if neighbor not in visited:
                    queue.append(neighbor)
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}
bfs_with_queue(graph, 'A')
```
