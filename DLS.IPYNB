```python
def dls(graph, start, target, limit, path=None):
    if path is None:
        path = []

    path.append(start)

    if start == target:
        return path

    if limit <= 0:
        path.pop()   
        return None

    for neighbor in graph[start]:
        result = dls(graph, neighbor, target, limit - 1, path)
        if result:
            return result

    path.pop()  
    return None



if __name__ == "__main__":
    graph = {
        "A": ["B", "C"],
        "B": ["D", "E"],
        "C": ["F"],
        "D": [],
        "E": ["F"],
        "F": []
    }

    start_node = "A"
    target_node = "F"
    depth_limit = 2

    print(f"DLS (limit={depth_limit}):")
    path = dls(graph, start_node, target_node, depth_limit)

    if path:
        print(f"Target '{target_node}' found within depth {depth_limit}")
        print("Path explored:", " -> ".join(path))
    else:
        print(f"Target '{target_node}' NOT found within depth {depth_limit}")
```
