```python
class MovieBudgetAnalyzer:
    def __init__(self, movies):
        self.movies = movies

    def calculate_average_budget(self):
        total_budget = sum(budget for _, budget in self.movies)
        return total_budget / len(self.movies)

    def show_high_budget_movies(self):
        average = self.calculate_average_budget()
        print(f"\nğŸ¬ Average Hollywood Movie Budget: ${average:,.0f}\n")

        count = 0
        print("ğŸ’° Movies with Higher than Average Budget:\n")
        for name, budget in self.movies:
            if budget > average:
                difference = budget - average
                print(f"ğŸ¥ {name} â€” ${budget:,.0f} (â†‘ ${difference:,.0f})")
                count += 1
        print(f"\nğŸ“Š Total movies above average budget: {count}")

    def add_movies(self):
        num = int(input("ğŸï¸ How many Hollywood movies would you like to add? "))
        for _ in range(num):
            name = input("ğŸ¬ Enter movie title: ")
            budget = int(input("ğŸ’µ Enter movie budget in USD: "))
            self.movies.append((name, budget))
        print("\nâœ… Movies added successfully!\n")

movies = [
    ("Inception", 160000000),
    ("The Dark Knight", 185000000),
    ("Avengers: Endgame", 356000000),
    ("Titanic", 200000000),
    ("Avatar", 237000000),
    ("Interstellar", 165000000),
    ("The Matrix", 63000000)
]

analyzer = MovieBudgetAnalyzer(movies)
analyzer.show_high_budget_movies()
```
