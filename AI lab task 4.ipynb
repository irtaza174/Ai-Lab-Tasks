```python
class Cleaner:
    def __init__(self, text):
        self.text = text
        self.punctuations = '''!()-[]{};:'"\\,<>./?@#$%^&*_~'''

    def remove_punctuation(self):
        clean_text = ""
        for char in self.text:
            if char not in self.punctuations:
                clean_text += char
        return clean_text

print("ðŸ‘‹ Hello! I'm your text cleaner.")
user_text = input("ðŸ“ Please enter the text you'd like me to tidy up: ")

cleaner = Cleaner(user_text)
result = cleaner.remove_punctuation()

print("\nâœ… Here's your cleaned-up text, free of punctuation:")
print(result)
```

```python
def bubble_sort(lst):
    n = len(lst)
    for i in range(n):
        for j in range(0, n - i - 1):
            if lst[j] > lst[j + 1]:
                lst[j], lst[j + 1] = lst[j + 1], lst[j]
    return lst

print("ðŸ‘‹ Welcome to the Bubble Sort Text Organizer!")
text = input("ðŸ“ Please enter a sentence or phrase: ")

words = text.split()
bubble_sorted_words = bubble_sort(words.copy())
print("\nðŸ”¤ Sorted by words (ASCII order):")
print("ðŸ§©", " ".join(bubble_sorted_words))

chars = list(text)
bubble_sorted_chars = bubble_sort(chars.copy())
print("\nðŸ”¡ Sorted by characters (including spaces):")
print("ðŸ§©", "".join(bubble_sorted_chars))

chars_no_space = list(text.replace(" ", ""))
bubble_sorted_chars_no_space = bubble_sort(chars_no_space.copy())
print("\nðŸ”¡ Sorted by characters (excluding spaces):")
print("ðŸ§©", "".join(bubble_sorted_chars_no_space))
```

```python
class LuhnValidator:
    def __init__(self, card_number):
        self.card_number = str(card_number)

    def is_valid(self):
        digits = [int(d) for d in self.card_number if d.isdigit()]
        checksum = 0
        parity = len(digits) % 2
        for i, digit in enumerate(digits):
            if i % 2 == parity:
                digit *= 2
                if digit > 9:
                    digit -= 9
            checksum += digit
        return checksum % 10 == 0
    
card = input("Enter card number: ")
validator = LuhnValidator(card)
if validator.is_valid():
    print("Valid card number.")
else:
    print("Invalid card number.")
```
