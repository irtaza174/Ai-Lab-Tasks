```python
class FileManager:
    def __init__(self, filename):
        """Initialize the file manager with a filename."""
        self.filename = filename

    def create_file(self):
        """Create a new empty file (or overwrite if exists)."""
        try:
            with open(self.filename, 'w') as file:
                pass
            print(f"‚úÖ File '{self.filename}' created successfully.")
        except Exception as e:
            print(f"‚ùå Error creating file: {e}")

    def write_to_file(self, content):
        """Write content to the file (overwrites existing content)."""
        try:
            with open(self.filename, 'w') as file:
                file.write(content)
            print(f"‚úçÔ∏è Written to '{self.filename}':\n{content}")
        except Exception as e:
            print(f"‚ùå Error writing to file: {e}")

    def append_to_file(self, content):
        """Append content to the file without removing existing data."""
        try:
            with open(self.filename, 'a') as file:
                file.write(content)
            print(f"‚ûï Appended to '{self.filename}':\n{content}")
        except Exception as e:
            print(f"‚ùå Error appending to file: {e}")

    def read_file(self):
        """Read and display the file content."""
        try:
            with open(self.filename, 'r') as file:
                content = file.read()
            print(f"üìñ Content of '{self.filename}':\n{content}")
            return content
        except FileNotFoundError:
            print(f"‚ö†Ô∏è File '{self.filename}' does not exist.")
        except Exception as e:
            print(f"‚ùå Error reading file: {e}")

    def delete_file_content(self):
        """Clear the file content without deleting the file."""
        try:
            with open(self.filename, 'w') as file:
                file.write('')
            print(f"üßπ Cleared content of '{self.filename}'.")
        except Exception as e:
            print(f"‚ùå Error clearing file: {e}")


# üß™ --- Example Usage ---
if __name__ == "__main__":
    # Create an object of FileManager for "myfile.txt"
    file_manager = FileManager("myfile.txt")

    file_manager.create_file()               # Create file
    file_manager.write_to_file("Hello!\n")   # Write new content
    file_manager.append_to_file("This is appended.\n")  # Append content
    file_manager.read_file()                 # Read and display content
    file_manager.delete_file_content()       # Clear file content
    file_manager.read_file()                 # Check if cleared
```
