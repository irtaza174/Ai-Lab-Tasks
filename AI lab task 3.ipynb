```python
class SimpleHeaterAgent:
    def __init__(self, setpoint):
        self.setpoint = setpoint
        self.last_action = "OFF"
        self.history = []

    def control_heater(self, temp):
        if temp < self.setpoint - 1:
            if self.last_action != "ON":
                self.last_action = "ON"
                action = "Turning the heater ON – it's getting chilly!"
            else:
                action = "Heater stays ON – still below comfort level."
        elif temp > self.setpoint + 1:
            if self.last_action != "OFF":
                self.last_action = "OFF"
                action = "Turning the heater OFF – it's warm enough."
            else:
                action = "Heater remains OFF – no need for extra heat."
        else:
            action = f"No change – keeping heater {self.last_action}."

        self.history.append((temp, action))
        return action

    def show_history(self):
        print("\n Heater Action Log:")
        for temp, action in self.history:
            print(f" Temp: {temp}°C → {action}")

agent = SimpleHeaterAgent(setpoint=22)
temps = [28, 21, 22, 24, 22, 21, 20, 19]

print(" Heater Agent Decisions:")
for t in temps:
    print(f" Temp: {t}°C → {agent.control_heater(t)}")

agent.show_history()
```
